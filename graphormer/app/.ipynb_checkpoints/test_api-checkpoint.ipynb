```python
import requests
from pydantic import BaseModel
class Points(BaseModel):
    start_lat: float = 73.244475	
    start_lon: float = 54.988728
    end_lat: float = 73.246867
    end_lon: float = 54.988618

a = Points()
body = {"start_lat": a.start_lat,
                    "start_lon": a.start_lon,
                    "end_lat": a.end_lat,
                    "end_lon": a.end_lon,
                   "type": 'lolo'}

r = requests.post('http://0.0.0.0:3006/get_path', headers = {'Content-Type': 'application/json'}, json = body)
weights_abakan = r.json()['abakan']
weights_omsk = r.json()['omsk']
```
