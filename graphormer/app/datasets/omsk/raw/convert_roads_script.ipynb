```python
import numpy as np
import json
import pathlib
import pandas as pd 

link = pathlib.Path().resolve()
link = str(link).split('TransTTE')[0]
global_root = link + 'TransTTE'

data_merge = pd.read_pickle(global_root + '/datasets/omsk/raw/omsk_full_routes_deeptte_states.pkl')
data_train = pd.read_csv(global_root + '/datasets/omsk/raw/omsk_full_routes_final_weather_L_NaN_filtered_FIXED.csv')
data_train['edges_geo'] = data_merge['edges_geo']
data_train['edges_int'] = data_merge['edges_int']

convert_table = data_train.copy()
convert_table = convert_table.dropna().reset_index(drop = True)

convert_table_valid = pd.DataFrame(columns = ['edge_id_start','edge_id_end','edge_coord_start','edge_coord_end'])
for i in range(0, convert_table.shape[0]):
    convert_table_itter = pd.DataFrame([[np.nan,np.nan,np.nan,np.nan]],columns = ['edge_id_start','edge_id_end','edge_coord_start','edge_coord_end'])
    convert_table_itter['edge_id_start'] = convert_table.loc[i]['edges_int'][0]
    convert_table_itter['edge_id_end'] = convert_table.loc[i]['edges_int'][-1]
    convert_table_itter['edge_coord_start'] = convert_table_itter['edge_coord_start'].astype(object)
    convert_table_itter['edge_coord_end'] = convert_table_itter['edge_coord_end'].astype(object)
    convert_table_itter.at[0, 'edge_coord_start'] = [list(convert_table.loc[i]['edges_geo'][0])]
    convert_table_itter.at[0, 'edge_coord_end'] = [list(convert_table.loc[i]['edges_geo'][-1])]
    convert_table_valid = pd.concat([convert_table_valid,convert_table_itter])
```

```python
convert_table_valid = convert_table_valid.reset_index(drop = True)
convert_table_valid = pd.merge(convert_table.reset_index(), convert_table_valid.reset_index(), on = ['index'], how = 'left')
```

```python
data = pd.read_csv('omsk_full_routes_final_weather_L_NaN_filtered_FIXED.csv')
data['edges_geo'] = convert_table_valid['edges_geo']
data['edges_int'] = convert_table_valid['edges_int']
data['edge_id_start'] = convert_table_valid['edge_id_start']
data['edge_id_end'] = convert_table_valid['edge_id_end']
data['edge_coord_start'] = convert_table_valid['edge_coord_start']
data['edge_coord_end'] = convert_table_valid['edge_coord_end']
```

```python
import pickle
with open('final_omsk.pickle', 'wb') as f:
    pickle.dump(data, f)
```
